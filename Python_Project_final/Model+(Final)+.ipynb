```python
import pandas as pd
import pickle

from sklearn.model_selection import cross_validate
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

import pandas as pd
df_train = pd.read_csv("df_cleaned.csv")

X_train_t = df_train.iloc[:,:72]
Y_train = df_train["RiskPerformance"]
```

## LDA

```python
cv_results_lda = cross_validate(LinearDiscriminantAnalysis(shrinkage = 0, solver = 'lsqr'), X_train_t, Y_train, cv=5, return_estimator=True)


(cv_results_lda['test_score']).mean()
```


## Log penalty l1

```python
from sklearn.linear_model import LogisticRegression
```

```python
cv_results_logl1 = cross_validate(LogisticRegression(max_iter = 10000, penalty = 'l1', solver = 'saga',
                                                  tol = 0.0001, random_state = 1), X_train_t, Y_train, cv=5, return_estimator=True)
```

```python
(cv_results_logl1['test_score']).mean()
```

## Log penalty none

```python
cv_results_logl2 = cross_validate(LogisticRegression(max_iter = 10000, penalty = 'none',
                                                  tol = 0.0001, random_state = 1), X_train_t, Y_train, cv=5, return_estimator=True)
```

```python
(cv_results_logl2['test_score']).mean()
```

# Tree

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import AdaBoostClassifier
```

## decision tree

```python
cv_results_DT = cross_validate(DecisionTreeClassifier(max_depth = 6, max_leaf_nodes = 100,
                                                  min_samples_leaf = 100, random_state = 1), X_train_t, Y_train, cv=5, return_estimator=True)
```

```python
(cv_results_DT['test_score']).mean()
```

## random forest

```python
cv_results_RF = cross_validate(RandomForestClassifier(max_depth = 7, max_leaf_nodes = 16,
                                                  min_samples_leaf = 20, n_estimators = 50 ,random_state = 1), X_train_t, Y_train, cv=5, return_estimator=True)
```

```python
(cv_results_RF['test_score']).mean()
```

## boosting log

```python
cv_results_logboosting = cross_validate(AdaBoostClassifier(LogisticRegression(), learning_rate = 1, n_estimators = 100,
                                                  random_state = 1), X_train_t, Y_train, cv=5, return_estimator=True)
```

```python
(cv_results_logboosting['test_score']).mean()
```

## bagging


```python
cv_results_bagging = cross_validate(BaggingClassifier(oob_score = 'True', n_estimators = 30,
                                                  random_state = 1), X_train_t, Y_train, cv=5, return_estimator=True)
```

```python
(cv_results_bagging['test_score']).mean()
```

## bagging log

```python
#bagging log
cv_results_baggingLog = cross_validate(BaggingClassifier(LogisticRegression(),oob_score = 'True', n_estimators = 30,
                                                  random_state = 1), X_train_t, Y_train, cv=5, return_estimator=True)
```

```python
(cv_results_baggingLog['test_score']).mean()
```

#### boosting log presents the best accuracy from this document, so we will retrain the model with the whole dataset

```python
model = AdaBoostClassifier(LogisticRegression(), learning_rate = 1, n_estimators = 100, random_state = 1).fit(X_train_t, Y_train)
```

# Output model and CSV file


```python
filename = 'finalized_model.sav'
with open(filename,'wb') as fp:
    pickle.dump(model,fp)
```
